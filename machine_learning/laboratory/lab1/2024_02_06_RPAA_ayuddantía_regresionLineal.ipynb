#Regresi√≥n lineal


```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression



df = pd.read_csv('salaries.csv')
print(df.describe)
exp=df['experiencia']
salario=df['salario']

x = np.array(exp).reshape(-1, 1)
y = np.array(salario)

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=.3, random_state=0)


regressor = LinearRegression()
regressor.fit(xtrain, ytrain)

pred = regressor.predict(xtest)


# x = np.linspace(0, 12, 100)  # Generar 100 puntos entre -5 y 5
# m = 8433.909212174  # Pendiente
# b = 26833.427381195  # Intercepto
# y = m * x + b

plt.scatter(xtrain, ytrain, color= 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
# plt.plot(x, y, '-g')

plt.title ("Training - Regresi√≥n lineal")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Salario")
plt.show()



plt.scatter(xtest, ytest, color= 'red')
plt.plot(xtest, pred, color = 'blue')
plt.title("Test - Regresi√≥n lineal")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Salario")
plt.show()



```

```python
mse = mean_squared_error(ytest, pred)
r2 = r2_score(ytest, pred)
mae = mean_absolute_error(ytest, pred)
rmse = np.sqrt(mse)

# m√©tricas para la regresi√≥n lineal
print("M√©tricas de evaluaci√≥n del modelo de regresi√≥n lineal:")
print("MSE:", mse)
print("R2:", r2)
print("MAE:", mae)
print("RMSE:", rmse)
```

# Red neuronal


```python
import tensorflow as tf
from sklearn.model_selection import train_test_split


xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.3, random_state=0)


model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)  # Una neurona y activaci√≥n lineal
])


model.compile(optimizer='sgd', loss='mean_squared_error')  # SGD es el optimizador, mean_squared_error es la funci√≥n de p√©rdida

model.fit(xtrain, ytrain, epochs=100, verbose=1)

loss = model.evaluate(xtest, ytest)
print("Test Loss:", loss)

predictions = model.predict(xtest)


plt.scatter(xtrain, ytrain, color= 'red')
plt.plot(xtrain, model.predict(xtrain), color = 'blue')
plt.title ("Training - Red neuronal")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Salario")
plt.show()

plt.scatter(xtest, ytest, color='red')
plt.plot(xtest, predictions, color='blue')
plt.title("Test - Red neuronal")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Salario")
plt.show()

print(model.evaluate(xtest,ytest))
```

```python
mse_nn = mean_squared_error(ytest, predictions)
r2_nn = r2_score(ytest, predictions)
mae_nn = mean_absolute_error(ytest, predictions)
rmse_nn = np.sqrt(mse_nn)

# m√©tricas para la red neuronal
print("M√©tricas de evaluaci√≥n del modelo de red neuronal:")
print("MSE:", mse_nn)
print("R2:", r2_nn)
print("MAE:", mae_nn)
print("RMSE:", rmse_nn)
```
