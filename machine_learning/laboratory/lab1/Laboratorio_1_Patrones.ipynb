## Pandas

```python
import pandas as pd #Importamos pandas
# Creamos un dataframe a partir de un diccionario con Nombre, Edad y Calificacion
data = {'Nombre': ['Juan', 'María', 'Pedro', 'Ana'],
        'Edad': [20, 21, 19, 22],
        'Calificación': [85, 90, 88, 92]}
df = pd.DataFrame(data)
df
```

```python
# Obtener la media de las edades
df["Edad"].mean()
```

```python
# Obtener la calificacion mas alta
df["Calificación"].max()
```

```python
# Filtrar por calificaciones mayores o iguales a 85
df[df["Calificación"]>85]
```

```python
# Leer el csv de articulos
df_articulos = pd.read_csv("articulos.csv")
df_articulos
```

```python
# Leer el csv de clasificaciones_articulos
df_clasificaciones = pd.read_csv("clasificaciones_articulos.csv"
)
df_clasificaciones
```

```python
#Hacer merge entre el csv de articulos y de clasificaciones .merge(left_on=,rigton=)
df_total = df_articulos.merge(df_clasificaciones,left_on="nombre",
                  right_on="nombre" )
df_total
```

```python
#Quitar valores nulos .fillna()
df_total = df_total.fillna(0)
df_total
```

```python
# Obtener el valor total de cada producto multiplicando las columnas de clasificacion y precio
df_total["total"] = df_total["cantidad"]*df_total["precio"]
df_total
```

```python
# Obtener el valor total por clasificacion .groupby
df = df_total.groupby("clasificacion")["total"].sum()
df
```

```python
#Guardar los resultados del total por clasificacion en un csv
df.to_csv("resultados.csv")
```

## Numpy

```python
# numpy
import numpy as np
#Creamos matrices
A = np.array([[1, 2], [3, 4]])
print(A)
B = np.array([[5, 6], [7, 8]])
B
```

```python
# Operaciones de matrices
print("A:")
print(A)

print("B:")
print(B)
print()

print("Suma de A y B")
suma = A + B
print(suma)
print()
# Resta de matrices
print("Resta de A y B")
resta = A - B
print(resta )
print()
# Multiplicacion de matrices (producto punto) .dot
print("Multiplicacion de A y B")
producto = A.dot(B)
print(producto )
print()

# Transposicion de una matriz .T
print("Transpuesta de Matriz A")
transpuesta_A = A.T
print(transpuesta_A)
print()
# Matriz cuadrada de ejemplo
matriz_cuadrada = np.array([[1, 2], [3, 4]])
print("La matriz es")
print(matriz_cuadrada)

# Calcular el determinante de la matriz .linalg.det
determinante = np.linalg.det(matriz_cuadrada)
print("El determinante de la matriz es:", determinante)
```

```python
# Sistema de ecuaciones:
# 2x + y = 5
# 3x - 2y = -8

# Coeficientes de las variables
coeficientes = np.array([[2, 1], [3, -2]])

# Términos constantes
terminos_constantes = np.array([5, -8])


# Solucionar el sistema .linalg.solve
solucion = np.linalg.solve(coeficientes,terminos_constantes)
print("La solución del sistema de ecuaciones es:", solucion)
```

```python
print("La matriz es:")
print(matriz_cuadrada)

# Calcular la inversa de la matriz np.linalg.inv
inversa = np.linalg.inv(matriz_cuadrada)

print("La inversa de la matriz es:")

print(inversa)
matriz_cuadrada.dot(inversa)
```

```python
# Matriz cuadrada de ejemplo
matriz_cuadrada = np.array([[1, 2], [3, 4]])
print("La matriz es:")
print(matriz_cuadrada)
# Calcular el polinomio característico np.poly()
polinomio_caracteristico = np.poly(matriz_cuadrada)
# DET(I-A(x))
print("El polinomio característico es:", polinomio_caracteristico)
```

```python
#Eigen valores y eigen vectores np.linalg.eig
eigenvalores, eigenvectores =np.linalg.eig(matriz_cuadrada)
print("Los eigenvalores son:", eigenvalores)
print("Los eigenvectores son:")
print(eigenvectores)
```

```python
eigenvectores
```

```python
#Interaccion entre numpy y pandas
print("DataFrame de eigenvectores:")
df = pd.DataFrame(eigenvectores)
print(df)
print()
#Volver a leer el dataframe de articulos con pandas y convertirlo a un array de numpy

print("DataFrame de articulos:")
print(df_articulos)
print()
array_articulos = np.array(df_articulos)
print("Array de articulos:")
print(array_articulos)
```

```python
#Guardar los eigenvectores
df.to_csv("eigenvectores.csv")
```

```python
# Generar muestras aleatorias de una distribución normal (gaussiana) np.random.normal(loc=,scale=,size=)
media = 0
desviacion_estandar = 1
muestras = np.random.normal(loc=media,scale=desviacion_estandar,size=10000)
print(muestras)
# Calcular la media y la desviación estándar de las muestras np.mean() y np.std()

media_muestras = np.mean(muestras)
desviacion_estandar_muestras = np.std(muestras)

print("Media de las muestras:", media_muestras)
print("Desviación estándar de las muestras:", desviacion_estandar_muestras)
```

```python
from sklearn.linear_model import LinearRegression
#Leer el csv de precios_casas
df_casas = pd.read_c<<sv("precios_casas.csv")
df_casas
```

```python
# obtener la lista de precios y de dimension, el de dimension convertirlo a array de numpy
tam = np.array(df_casas["tam"])
print(tam)
precios = np.array(df_casas["precios"])
precios
```

```python
#Aplicar reshape() al array de longitudes para transponerlo
tam = tam.reshape(-1,1)
tam
```

```python
# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo con los datos
modelo.fit(tam,precios) #Fit entrena
# Predecir el precio de una casa de 1300 pies cuadrados
modelo.predict(np.array([[1300]]))

# print("El precio predicho de una casa de 1300 pies cuadrados es:", precio_predicho[0])


```

```python
# Graficar los datos y la línea de regresión
plt.scatter(tam, precios, color='blue')
plt.plot(tam, modelo.predict(tam), color='red')
plt.title('Regresión Lineal: Precio de Casas vs Tamaño')
plt.xlabel('Tamaño de la Casa (pies cuadrados)')
plt.ylabel('Precio de la Casa (dólares)')
plt.show()
```
